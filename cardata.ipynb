```python
df1 =df[df['Selling_Price']<30]
df1
```

```python
sns.scatterplot(data = df1, x ='Transmission' , y = 'Selling_Price' )
```

```python
df1.loc[:, 'Fuel_Type'] = df1['Fuel_Type'].replace({"Petrol":2, "Diesel":3, "CNG":4})
pd.set_option('future.no_silent_downcasting', True)  
df1.loc[:, 'Seller_Type'] = df1['Seller_Type'].replace({"Dealer": 2, "Individual": 3})  
pd.set_option('future.no_silent_downcasting', True)  
df1.loc[:, 'Transmission'] = df1['Transmission'].replace({"Manual": 2, "Automatic": 3})  
df1
```

```python
X = pd.DataFrame(df1 ,columns =['age','Present_Price','Kms_Driven','Fuel_Type','Seller_Type','Transmission','Owner'])
Y = df1['Selling_Price'].values.reshape(-1,1)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size=0.3 , random_state=0)
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(X_train ,Y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
kfold_validation = KFold(6)
```

```python
result = cross_val_score(regressor , X,Y , cv = kfold_validation)
```

```python
print(result)
print(np.mean(result))
```
