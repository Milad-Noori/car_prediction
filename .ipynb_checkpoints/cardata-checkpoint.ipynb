```python
df1 =df[df['Selling_Price']<30]
df1
```

```python
sns.scatterplot(data = df1, x ='Transmission' , y = 'Selling_Price' )
```

```python
print(pd.unique(df1['Fuel_Type']))
print(pd.unique(df1['Seller_Type']))
print(pd.unique(df1['Transmission']))
print(pd.unique(df1['Owner']))
```

```python
df1['Fuel_Type'].replace({'Petrol':2, 'Diesel':3, 'CNG': 4} , inplace=True)
df1['Seller_Type'].replace({'Dealer':2, 'Individual':3},inplace=True)
df1['Transmission'].replace({'Manual':2, 'Automatic':3},inplace=True)
df1
```

```python
df1.corr()
```

```python
X = pd.DataFrame(df1 ,columns =['age','Present_Price','Kms_Driven','Fuel_Type','Seller_Type','Transmission','Owner'])
Y = df1['Selling_Price'].values.reshape(-1 , 1) 
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size=0.3 , random_state=0)
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(X_train ,Y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
kfold_validation = KFold(10)
```

```python
result = cross_val_score(regressor , X,Y , cv = kfold_validation)
```

```python
print(result)
print(np.mean(result))
```

```python
data1 = df[df.index<150]
data2 = df[df.index>200]
data3 = pd.concat([data1 , data2] )
data3.reset_index(drop =True )
data3
```

```python
X2 = data3.drop('Selling_Price',axis = 1)
Y2 = data3['Selling_Price']
```

```python
from sklearn.preprocessing import LabelEncoder  


label_encoder = LabelEncoder()  
for col in X2.select_dtypes(include=['object']).columns:  
    X2[col] = label_encoder.fit_transform(X2[col])  
```

```python
X_train2 , X_test2 , Y_train2 , Y_test2 = train_test_split(X2,Y2 , test_size=0.3 , random_state=0) 
regressor1 = LinearRegression()
regressor1.fit(X_train2 , Y_train2 )
```

```python
y_pred2=regressor1.predict(X_test2)
```

```python
print("MAE IS :", metrics.mean_absolute_error(Y_test2 , y_pred2))
print("MSE IS :" ,metrics.mean_squared_error(Y_test2 , y_pred2))
print("RMSE IS :",np.sqrt(metrics.mean_squared_error(Y_test2 , y_pred2)))
print("R2score :" , metrics.r2_score(Y_test2 ,y_pred2))
```

```python
plt.scatter(X_test2['Present_Price'], Y_test2 , color="blue" )####  واقعی
plt.scatter(X_test2['Present_Price'], y_pred2  , color="red" )####  پیش بینی شده
```
